# 2. Bootstrap untuk menguji hipotesis

```python
#mengambil data berdasarkan kategori yang diinginkan
baik = data[data['Account.Balance']==4]
buruk = data[data['Account.Balance']!=4]

#mengambil data dari variabel yang akan diuji
baik1 =baik['Creditability']
buruk1 = baik['Creditability']

good= np.array(baik1)
bad = np.array(buruk1)


#melakukan resampling
def bootstrap_resample(A, n=None):
    if n == None:
        n = len(A)
    np.random.seed(8)    
    resample_i = np.floor(np.random.rand(n)*len(A)).astype(int)
    resample = A[resample_i]
    return resample

baik_resample = bootstrap_resample(good, n=10000)
buruk_resample = bootstrap_resample(bad, n=10000)
```

```python
good_bootstrap=[]
for i in range(10000):
    result_resample = bootstrap_resample(baik_resample,n=363)
    m=result_resample.mean()
    good_bootstrap.append(m)
good_bootstrap = np.array(good_bootstrap)
good_bootstrap.mean()
```

```python
bad_bootstrap=[]
for i in range(10000):
    result_resample = bootstrap_resample(buruk_resample,n=537)
    m=result_resample.mean()
    bad_bootstrap.append(m)
bad_bootstrap = np.array(bad_bootstrap)
bad_bootstrap.mean()
```

```python
import scipy.stats as ss
ss.ttest_ind(good_bootstrap,bad_bootstrap)
```

Pada statistika inferensia ini akan dibuktikan bahwa orang-orang yang mempunyai Saldo (Account.Balance) â‰¥ 200 USD memiliki tingkat credibility 25% lebih tinggi daripada orang lainnya (tidak punya akun, tidak punya saldo dan saldo kurang dari 200 USD) menggunakan metode resampling dengan hipotesis sebagai berikut
H0:orang yang mempunyai saldo => 200 USD memiliki tingkat credibility 25% <= dari orang lainnya
H1:orang yang mempunyai saldo => 200 USD memiliki tingkat credibility 25% > dari orang lainnya

Statistika uji: P-value (0.0) < alfa (0.05)
keputusan: tolak H0
Kesimpulan: terbukti bahwa orang-orang yang mempunyai saldo >=200USDmemiliki tigkat credibility 25% lebih besar dari yang lainnya

# 3. Statistika Inferensia Untuk Menguji Hipotesis

Pada bagian ini saya akan menguji apakah dua populasi baik pada variabel duration of credit, credit amount, dan age memiliki perbedaan untuk pelanggan dengan status credit worthy maupun not-credit worthy

```python
x1 = data[data['Creditability'] == 0]['Duration.of.Credit']
x2 = data[data['Creditability'] == 1]['Duration.of.Credit']
import numpy as np
import statsmodels.api as sm
from scipy import stats
import statsmodels.formula.api as smf
sm.stats.ztest(x1, x2, value=0, alternative='two-sided', usevar='pooled', ddof=1.0)
```

Untuk pengujian pertama digunakan z test karena jumlah data yang cukup banyak. dari pengujian ini kita akan mencari tau apakah rata-rata Durasi kredit pelanggan dengan status baik sama dengan rata-rata Durasi kredit pelanggan dengan status buruk dengan menggunakan hipotesis

H0: rata-rata Durasi kredit pelanggan berstatus baik = rata-rata Durasi kredit pelanggan berstatus buruk
H1: rata-rata Durasi kredit pelanggan berstatus baik != rata-rata Durasi kredit pelanggan berstatus buruk

Uji statistik : nilai pvalue= 2.365 x 10^-12<alpha=0.05

Keputusan: Tolak H0
Kesimpulan: Rata-rata Durasi kredit pelanggan berstatus baik tidak sama dengan rata-rata Durasi kredit pelanggan berstatus buruk


```python
x1 = data[data['Creditability'] == 0]['Credit.Amount']
x2 = data[data['Creditability'] == 1]['Credit.Amount']
import numpy as np
import statsmodels.api as sm
from scipy import stats
import statsmodels.formula.api as smf
sm.stats.ztest(x1, x2, value=0, alternative='two-sided', usevar='pooled', ddof=1.0)
```

Pada pengujian ini kita akan mencari tau apakah rata-rata Jumlah Kredit pelanggan dengan status baik sama dengan rata-rata Jumlah Kredit pelanggan dengan status buruk dengan menggunakan hipotesis

H1: rata-rata Jumlah Kredit pelanggan berstatus baik != rata-rata Jumlah Kredit pelanggan berstatus buruk

Uji statistik : nilai pvalue= 2.77 x 10^-7<alpha=0.05

Keputusan: Tolak H0
Kesimpulan: Rata-rata untuk Jumlah Kredit pelanggan berstatus baik tidak sama dengan rata-rata Jumlah Kredit pelanggan berstatus buruk

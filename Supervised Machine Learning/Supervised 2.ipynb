# 1. How value of 'b' affect 'success' probability?

### 1.1 With b>0 

```python
#define function with b>0

def reglog_pos(x):
    return math.exp(0.01+0.8*x)/(1+math.exp(0.01+0.8*x))
```

```python
reglog_pos(1)
```

```python
func_pos=np.vectorize(reglog_pos)
```

# Example 1

```python
x1= np.random.randn(1000)
```

```python
plt.hist(x1, color='pink')
plt.show()
```

```python
pi_pos=func_pos(x1)
```

```python
plt.scatter(x1,pi_pos,color='pink')
plt.show()
```

peningkatan x diikuti dengan peningkatan nilai p(x)

# Example 2

```python
x2=np.random.uniform(-10,10,1000)
```

```python
plt.hist(x2,color='pink')
plt.show()
```

```python
pi_pos=func_pos(x2)
```

```python
plt.scatter(x2,pi_pos,color='pink')
plt.show()
```

Berapapun nilai x nya, peluangnya tetap 0-1

## 1.2. With b<0 

```python
#define function with b<0

def reglog_neg(x):
    return math.exp(0.01+(-0.8)*x)/(1+math.exp(0.01+(-0.8)*x))
```

```python
func_neg=np.vectorize(reglog_neg)
```

# Example 1

```python
x1=np.random.randn(1000)
```

```python
plt.hist(x1,color='pink')
plt.show()
```

```python
pi_neg=func_neg(x1)
```

```python
plt.scatter(x1,pi_neg,color='pink')
plt.show()
```

```python
x2=np.random.uniform(-10,10,1000)
```

```python
plt.hist(x2,color='pink')
plt.show()
```

```python
pi_neg=func_neg(x2)
```

```python
plt.scatter(x2,pi_neg,color='pink')
plt.show()
```

#### ketika nilai b>0, maka nilai p(x) meningkat bersama dengan peningkatan nilai x
#### ketika nilai b<0, maka nilai p(x) menurun bersama dengan penurunan nilai x
#### ketia nilai b=0, maka nilai p(x) akan selalu konstan pada berapapun nilai xnya

# interpretasi
#### ODD= peluang sukses/peluang gagal
#### ODD= phi(x)/(1-phi(x))

#### jika y=1 untuk kejadian sukses dan y=0 untuk kejadian gagal
#### ODD=P(Y=1)/P(Y=0)




# Example 1

##### y=status pembayaran
##### x=tenorpeminjaman(bulan)
#### success event=terlambat bayar
#### Y=1 terlambat bayar
#### Y=0 tepat bayar

```python
x1=np.random.randn(1000)
pi_pos=func_pos(x1)

plt.scatter(x1,pi_pos)
plt.show()
```

```python
#probability 'success' when x=1

p_success=func_pos(1)
p_success
```

```python
#probability 'failure' when x=1
p_failure=1-func_pos(1)
p_failure
```

# ODD when x=1
# =P(success|X=1)/P(failure|X=1)

```python
# ODD when x=1
ODD=p_success/p_failure
ODD
```

#### Interpretasi: peluang sukses pada x=1 adalah 2.25 kali dibandingkan peluang gagal
    
#### peluang terlambat bayar jika tenor peminjaman 1 bulan adalah 2.25 kali peluang tepat bayar

## Missing Value

```python
import numpy as np
import pandas as pd
df=pd.DataFrame({'A':[1,2,np.nan],
                'B':[5,np.nan,np.nan],
                'C':[1,2,3]})
df
```

```python
#jika data yang digunakan sedikit sehingga lebih mudah dilihat
print(df.isnull())
#jika data yang digunakan banyak sehinga sulit diliat jadi langsung dicari saja jumlahnya
print(df.isnull().sum())
```

```python
#didrop barisnya, default axis=0
print(df.dropna())
#didrop kolomnya
print(df.dropna(axis=1))
```

 fungsi drop hanya digunakan untuk view aja sehingga jika nanti dipanggil lagi df bakal balik semua. kalo mau drop permanen bisa di save dengan nama variabel yang sama

```python
df.dropna(thresh=2)
```

membuang ketika dalam baris terdapat 2 atau lebih missing value

```python
#mengisi nilai null dengan tulisan FILLNA
print(df.fillna(value='FILL VALUE'))
#mengisi nilai null pada kolom A dengan nilai mean A
print(df['A'].fillna(value=df['A'].mean()))
```

```python
df.fillna(value=df.mean())
```

### metode fillna

```python
df.fillna(method='ffill')
```

```python
df.fillna(method='bfill')
```

## Merging, Joining, and Concatenating 

```python
import pandas as pd
df1=pd.DataFrame({'A':['A0','A1','A2','A3'],
                 'B':['B0','B1','B2','B3'],
                 'C':['C0','C1','C2','C3'],
                 'D':['D0','D1','D2','D3']},
                 index=[0,1,2,3])
df2=pd.DataFrame({'A':['A4','A5','A6','A7'],
                 'B':['B4','B5','B6','B7'],
                 'C':['C4','C5','C6','C7'],
                 'D':['D4','D5','D6','D7']},
                 index=[4,5,6,7])
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2])
```

buat digabungnya ke baris bawahnya

```python
pd.concat([df1,df2],axis=1)
```

digabung ke kolom baru

```python
left=pd.DataFrame({'key':['K0','K1','K2','K3'],
                   'A':['A0','A1','A2','A3'],
                   'B':['B0','B1','B2','B3']})
right=pd.DataFrame({'key':['K0','K1','K2','K3'],
                   'C':['C0','C1','C2','C3'],
                   'D':['D0','D1','D2','D3']})
```

```python
left
```

```python
right
```

### MERGE 

##### UNTUK DATA FRAME YANG JUMLAH KOLOM BARIS SAMA

```python
pd.merge(left,right,how='inner',on='key')
```

```python
pd.merge(left,right,how='outer',on='key')
```

```python
pd.merge(left,right,how='left',on='key')
```

```python
pd.merge(left,right,how='right',on='key')
```

bedanya merge sama join, kalo join cuma bisa pake indeks

```python
new=pd.DataFrame({'key':['K0','K1','K2'],
                   'C':['C0','C1','C2'],
                   'D':['D0','D1','D2']})
new
```

######  UNTUK YANG JUMLAH KOLOM  DAN BARISNYA BEDA

```python
pd.merge(left,new,how='outer',on='key')
```

```python
pd.merge(left,new,how='left',on='key')
```

```python
pd.merge(left,new,how='right',on='key')
```

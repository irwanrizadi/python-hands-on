# Visualization

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
a=np.random.rand(10,1)
b=np.random.rand(10,1)
c=np.random.rand(10,1)
print(a)
print(' ')
print(b)
print(' ')
print(c)
```

```python
#a sebagai x b sebagai y
plt.plot(a,b,'red') #untuk warna
plt.xlabel('x') #label untuk sumbu x
plt.ylabel('y')#label untuk sumbu y
plt.title('Random') #label untuk judul
plt.show() #menampilkan
```

```python
#a sebagai x b sebagai y
#untuk warna
plt.plot(a,b,'red')
plt.plot(a,c,'blue')
plt.xlabel('x') #label untuk sumbu x
plt.ylabel('y')#label untuk sumbu y
plt.title('Random') #label untuk judul
plt.show() #menampilkan
```

```python
plt.plot(a,b,color='#FFDD44') #hex color code (RRGGBB)
plt.plot(a,c,color=(1.0,0.2,0.3)) #RGB Tuple, bejrween 0-1
plt.xlabel('x') #label untuk sumbu x
plt.ylabel('y')#label untuk sumbu y
plt.title('Random') #label untuk judul
plt.show() 
```

```python
plt.annotate('red',xy=(0.1,0.2))#x coordinate x y koordinat y
plt.annotate('blue',xy=(0.5,0.7))
plt.show()
```

```python
plt.plot(a,b,'red', label = 'red')
plt.plot(a,c,'blue', label = 'blue')
plt.xlabel('y')
plt.title('Random')
plt.axis('tight')

#legend
plt.legend(loc = 'upper right') # upper right/left, lower right/left

#annotations
plt.annotate('red',xy=(0.1,0.2)) # (xy = (x coordinate, y coordinate))
plt.annotate('blue',xy=(0.5,0.7))
plt.show()
```

```python
plt.plot(a,b,'red', label = 'red')
plt.plot(a,c,'blue', label = 'blue')
plt.xlabel('y')
plt.title('Random')
plt.axis('tight')

#legend
plt.legend(loc = 'upper right') # upper right/left, lower right/left

#annotations
plt.annotate('red',xy=(0.1,0.2),xytext=(0.4,0.2),arrowprops={'color':'red'})
                                 # (xy = (x coordinate, y coordinate)) (palanya)
                                    #xytext starting position (buntutnya)
                                     #xy(arrow pointing to)
plt.annotate('blue',xy=(0.5,0.7),xytext=(0.55,0.75),arrowprops={'color':'blue'})
plt.show()
```

```python
#style
plt.plot(a,b,label = 'One')
plt.plot(a,c, label = 'Two')
plt.xlabel('y')
plt.title('Random')
plt.axis('tight')

#legend
plt.legend(loc = 'upper right') # upper right/left, lower right/left

#annotations
plt.annotate('red',xy=(0.1,0.2),xytext=(0.4,0.2),arrowprops={'color':'red'})
                                 # (xy = (x coordinate, y coordinate)) (palanya)
                                    #xytext starting position (buntutnya)
                                     #xy(arrow pointing to)
plt.annotate('blue',xy=(0.5,0.7),xytext=(0.55,0.75),arrowprops={'color':'blue'})
plt.style.use('ggplot')
plt.show()
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips)
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='day',palette='Set1')
plt.show()
```

seluruh data akan dibuat garis regresinya meskipun korelasinya rendah, jadi sebelum menginterpretasikan hasil di teliti dulu apakah korelasinya baik atau tidak. kemudian semakin kecil korelasinya maka akan sejajar dengan sumbu 
x

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',hue='day',palette='Set1')
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,row='sex',hue='day',palette='Set1')
plt.show()
```

```python
sns.stripplot(y='tip',data=tips)
plt.ylabel('tip($)')
plt.title('Tip Distribution')
plt.show()
```

```python
sns.stripplot(x='day',y='tip',data=tips, size=10, jitter=True)
plt.ylabel('tip($)')
plt.title('Tip Distribution')
plt.show()
```

jitter gunanya buat bikin titik titik nya ga cuma satu garis tapi numpuk-numpuk gitu

```python
sns.swarmplot(x='day',y='tip',data=tips)
plt.ylabel('tip($)')
plt.show()
```

```python
sns.swarmplot(x='tip',y='day',data=tips, hue='sex',palette='Set2',orient='h')
plt.ylabel('tip($)')
plt.axis('tight')
plt.show()
```

kalo bar chart karna kita bermain dengan kategorikal variabel, ketika x nya banyak jadi lebih baik horizontal yang di tampilkan untuk data tips karna kategoriknya cuma empat baiknya vertikal

```python
sns.boxplot(x='day',y='tip', data=tips, hue='sex',palette='Set2')
plt.ylabel('tip($)')
plt.tight_layout()
plt.show()
```

# 1. Join ALL

```python
d6=pd.merge(d5,d1,how='inner',on='OwnerID')
d6
```

```python
d7=pd.merge(d6,d2,how='inner',on=('ProcedureSubCode','ProcedureType'))
d7
```

# 2. Procedure type paling banyak

```python
y=d7.groupby('ProcedureType')['PetID'].count()
y.sort_values(ascending=False).head(1)
```

# 3.Owner yang kedokter>3

```python
own=d7.groupby(['OwnerID']).filter(lambda x:len(x)>3).groupby(['Name_y','Surname'])['PetID'].count()
own
```

# 4. Siapa yang paling sering ke dokter saat weekday

```python
wday=d7[(d7.Day!='Saturday')&(d7.Day!='Sunday')]
weekday=(wday.groupby('OwnerID')['OwnerID'].count().max())
wday.groupby(['OwnerID']).filter(lambda x:len(x)==weekday).groupby(['Name_y','Surname'])['PetID'].count()
```

# 5. Total Spending Jerry dan Jessica

```python
d7.groupby('Name_y')
j=d7[(d7.Name_y=='Jerry')|(d7.Name_y=='Jessica')].groupby(['Name_y','Surname'])['Price'].sum()
print(j)
print('total spending mereka adalah ', j[0]+j[1])
```

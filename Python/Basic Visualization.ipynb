```python
a=df2[["a","b"]].plot.area(alpha=0.5) #plot kolom a dan b
b=df2[["a","c"]].plot.area(alpha=0.5) #plot kolom a dan c
c=df2[["a","b"]].plot.area(alpha=0.7,figsize=(10,3)) #figsize buat ukuran grafik panjang x lebar
d=df2[["a","c"]].plot.area(alpha=0.7,figsize=(4,3)) 
print(a)
print(b)
print(c)
print(d)
```

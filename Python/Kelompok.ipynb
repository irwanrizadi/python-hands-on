## 1. Import Pets1, Pets2, & Owners Data

```python
import numpy as np
import pandas as pd
data1=pd.read_csv('D:/Owners.csv')
data2=pd.read_csv('D:/Pets1.csv')
data3=pd.read_csv('D:/Pets2.csv')
data4=pd.concat([data2,data3])
data4.head()
```

```python
data=pd.merge(data1,data4,how='left',on='OwnerID')
data.head()
```

```python
data.reset_index
```

```python
data[data.Kind=='Dog']=data.fillna(value={'Gender':'male'})
data[data.Kind=='Parrot']=data.fillna(value={'Gender':'male'})
data[data.Kind=='Cat']=data.fillna(value={'Gender':'female'})
data.head()
```

```python
a=data.groupby(['Kind','Gender'])['Age'].mean()
a
```

```python
data[(data.Kind=='Cat') & (data.Gender=='male')]= data.fillna(value={'Age':a[1]})
data[(data.Kind=='Cat') & (data.Gender=='female')]= data.fillna(value={'Age':a[0]})
data[(data.Kind=='Dog') & (data.Gender=='male')]= data.fillna(value={'Age':a[3]})
data[(data.Kind=='Dog') & (data.Gender=='female')]= data.fillna(value={'Age':a[2]})
data[(data.Kind=='Parrot') & (data.Gender=='male')]= data.fillna(value={'Age':a[5]})
data[(data.Kind=='Parrot') & (data.Gender=='female')]= data.fillna(value={'Age':a[4]})
data
```
